```python
from sklearn.cluster import KMeans
from matplotlib import cm
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.cluster import AgglomerativeClustering

import os
import pandas as pd
import numpy as np
import datetime as dt
import zipfile
import seaborn
import matplotlib.pylab as plt
import geopandas as gp

%matplotlib inline
```

```python
year = []
for i in range(94, 100, 1):
    year.append(str(i))
for i in range(0, 10, 1):
    year.append('0'+str(i))
for i in range(10, 15, 1):
    year.append(str(i))
```

```python
for i in year:
    os.system("mv " + "zbp" + i + "totals.zip " + os.getenv("PUIDATA"))

print ('zbp totals zip files moved to {}'.format(os.getenv("PUIDATA")))
```

# Getting NYC ZipCodes

```python
zip_codes = gp.read_file('nyc-zip-code-tabulation-areas-polygons.geojson')
zip_codes.head(3)
```

```python
zip_codes = zip_codes.drop([ u'@id', u'BLDGpostalCode', u'PO_NAME', u'CTY_FIPS', u'ST_FIPS',
                             u'STATE', u'OBJECTID', u'Shape_Area', u'borough', u'Shape_Leng'], axis = 1)
zip_codes = zip_codes.rename(columns={'postalCode' : 'zip_code'})
zip_codes['zip_code'] = pd.to_numeric(zip_codes['zip_code'], errors = 'coerce')
zip_codes.head()
```

### Clean NYC Business data. Year range is 1994 to 2014. Merge it with NYC Zip Code data

```python
for i in year:
    zf = zipfile.ZipFile(os.getenv("PUIDATA") + "/" + "zbp" + i + "totals.zip")
    temp = pd.read_csv(zf.open('zbp'+i+'totals.zip'.replace('.zip','.txt')))
    temp.columns = map(str.lower, temp.columns)
    temp = temp[['zip', 'est']]
    year = dt.datetime.strptime(str(i), '%y').year
    temp.rename(columns={'est': str(year), 'zip':'zip_code'}, inplace=True)
    zip_codes = zip_codes.merge(temp, on = 'zip_code', how = 'inner')
```

## Standardizing the time series

```python
ts_mean = zip_codes.iloc[:,range(2,23)].mean(axis = 1)
ts_std = zip_codes.iloc[:,range(2,23)].std(axis = 1)
zip_codes.iloc[:,range(2,23)] = zip_codes.iloc[:,range(2,23)].apply(lambda x: (x-ts_mean)/ts_std)
zip_codes.head(3)
```

```python
zip_codes.shape
```

```python
print 'After standardizing the time series - \n - Std. deviation along all zip codes = {a}\
\n - Mean along all zip codes = {b}(~0)'.format(
    a=zip_codes.iloc[:,range(2,23)].std(axis = 1).sum()/len(zip_codes),
    b=zip_codes.iloc[:,range(2,23)].mean(axis = 1).sum()/len(zip_codes))
```

# Clustering the data

```python
X = zip_codes.iloc[:, range(2,23)]
range_n_clusters = range(2,10)
for n_clusters in range_n_clusters:
    km = KMeans(n_clusters=n_clusters, random_state=324)
    cluster_labels = km.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters = {},".format(n_clusters)+" the average silhouette_score is {}".format(silhouette_avg))
```

As per the Silhouette score, the optimal number of clusters can be 2. 
Here, I have used 5 clusters for the analysis, as the silhoutte value has drastically reduced for 6 clusters (almost by 0.20).

### Clustering using K-means
### Number of Clusters = 2

```python
# K-Means cluster
kmeans_cluster2 = KMeans(2, random_state= 125).fit(zip_codes.iloc[:, range(2,23)])
centers2 = kmeans_cluster2.cluster_centers_
zip_codes['km2_labels'] = kmeans_cluster2.labels_
```

```python
fig = plt.figure(figsize=(20,8))
ax = fig.gca()
for i in range(len(zip_codes)):
    ax.plot(range(1994,2015), zip_codes.iloc[i,range(2,23)], 
            '-', lw = 1, alpha = 0.8, c=cm.hot(zip_codes.km2_labels[i]*50))

lines = range(len(centers2))    
for i in range(len(centers2)):
    line, = plt.plot(range(1994,2015),centers2[i], 
                    '.-', ms = 20, lw = 5, c=cm.hot(((i+1)*100)), 
                     label = 'cluster'+ str(i))
    lines[i] = (line)

plt.legend(handles=[lines[i] for i in range(len(centers2))], fontsize = 20, 
           loc = 'best')
plt.xlim(1994,2014)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("Years", fontsize=20)
plt.ylabel("Standardized values of establishments", fontsize=20)
plt.title('Time series of NYC Business establishments from 1994 to 2014 clustered into 2 groups', 
          fontsize=22)

plt.show()
```

#### Figure1: The above plot shows the time series of "No. of Business Establishments" in different zipcodes in NYC from 1994 to 2014. The time series of the zipcodes is standarized and clustered into 2 groups. 
#### The clustering seems to be as per the trend observed in the mean number of establishments in the clustered zipcodes.   

### Number of Clusters = 5

```python
# K-Means cluster
kmeans_cluster5 = KMeans(5, random_state= 12).fit(zip_codes.iloc[:, range(2,23)])
centers5 = kmeans_cluster5.cluster_centers_
zip_codes['km5_labels'] = kmeans_cluster5.labels_
```

```python
fig = plt.figure(figsize=(20,8))
ax = fig.gca()
for i in range(len(zip_codes)):
    ax.plot(range(1994,2015), zip_codes.iloc[i,range(2,23)], 
            '-', lw = 1, alpha = 0.3, c=cm.hot(zip_codes.km5_labels[i]*50))

lines = range(len(centers5))    
for i in range(len(centers5)):
    line, = plt.plot(range(1994,2015),centers5[i], 
                    '.-', ms = 20, lw = 5, c=cm.hot(((i+1)*80 - (i*35))), 
                     label = 'cluster'+ str(i))
    lines[i] = (line)

plt.legend(handles=[lines[i] for i in range(len(centers5))], fontsize = 15, 
           loc = 'best')
plt.xlim(1994,2014)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("Years", fontsize=20)
plt.ylabel("Standardized values of establishments", fontsize=20)
plt.title('Time series of NYC Business establishments from 1994 to 2014 clustered into 5 groups', 
          fontsize=22)

plt.show()
```

#### Figure2: The above plot shows the time series of "Number of Business Establishments" in different zipcodes in NYC from 1994 to 2014. The time series of the zipcodes is standarized and clustered into 5 groups. 
#### The clustering in this plot seems to be as per the trend observed in the mean number of establishments in the clustered zipcodes but at more granular level than it was with the 2 clusters as seen in Figure 1

## Cluster the data using Agglomerative Clustering

### Number of Clusters = 2

```python
agc2 = AgglomerativeClustering(n_clusters=2, 
                              compute_full_tree=True).fit(zip_codes.iloc[:, range(2,23)])
agg2_labels = agc2.labels_
zip_codes['agg2_labels'] = agc2.labels_
```

### Number of Clusters = 5

```python
agc5 = AgglomerativeClustering(n_clusters=5, 
                              compute_full_tree=True).fit(zip_codes.iloc[:, range(2,23)])
agg5_labels = agc5.labels_
zip_codes['agg5_labels'] = agc5.labels_
```

# Visualizing the clustered Zipcodes

```python
fig = plt.figure(figsize=(20,10))
ax1, ax2 = fig.add_subplot(121), fig.add_subplot(122)
zip_codes.plot(cmap =cm.viridis, column='agg2_labels', ax=ax1)
ax1.set_title("Agglomerative Clustering, k=2", fontsize=25)
ax1.grid()
ax1.set_xticks([])
ax1.set_yticks([])

zip_codes.plot(cmap =cm.viridis, column='km2_labels', ax=ax2)
ax2.set_title("K-means Clustering, k=2", fontsize=25)
ax2.grid()
ax2.set_xticks([])
ax2.set_yticks([])

plt.show()
```

#### Figure 3: Maps of clustered zip codes from the NYC business data. The clusters created by the K-means and Agglomerative clustering algorithms are almost same. 
#### The yellow part in the map above corresponds to the cluster 1 in Figure 1 i.e less increase in number of business establishments in areas around midtown, downtown manhattan and some areas in bronx, queens and brooklyn. 
#### This observation can be useful to further investigate the reason for increase in business in other areas; factors like reduction in crime can be analysed
